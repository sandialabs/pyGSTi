# "Automagic" parallel GST using MPI
This tutorial demonstrates how to compute GST estimates in parallel using MPI. 
This requires the `mpi4py` python package, which must be able to connect to an underlying MPI library (say, openmpi).

```python
import pygsti
from pygsti.modelpacks import smq1Q_XYI as mp
exp_design  = mp.create_gst_experiment_design(max_max_length=32)
mdl_datagen = mp.target_model().depolarize(op_noise=0.1, spam_noise=0.001)
data  = pygsti.data.simulate_data(mdl_datagen, exp_design.all_circuits_needing_data, num_samples=1000, seed=2020)
pdata = pygsti.protocols.ProtocolData(exp_design, data)
```

```python
protocol = pygsti.protocols.StandardGST("full TP")
results = protocol.run_mpi(
    pdata, num_workers=3, memlimit=2.1*(1024)**3,
    mpiexec='/opt/miniconda3/envs/leak311/bin/mpiexec'
)
report = pygsti.report.construct_standard_report(
    results, title="MPI Example Report", verbosity=2
)
report.write_html('example_files/mpi_example_brief', auto_open=True)
```
