# This notebook creates the serialized FOGI models for the modelpacks. So far only implemented for 1 qubit

```python
import pygsti
from pygsti.modelpacks import smq1Q_XY, smq1Q_XYI, smq1Q_XYZI, smq1Q_XZ, smq1Q_ZN

modelpacks = [smq1Q_XY, smq1Q_XYI, smq1Q_XYZI, smq1Q_XZ, smq1Q_ZN]
model_names = ['smq1Q_XY_GLND_FOGI.json', 'smq1Q_XYI_GLND_FOGI.json', 'smq1Q_XYZI_GLND_FOGI.json', 'smq1Q_XZ_GLND_FOGI.json', 'smq1Q_ZN_GLND_FOGI.json']

basis1q = pygsti.baseobjs.Basis.cast('pp', 4)
op_abbrevs = {('Gxpi2', 0): 'Gx',
				('Gypi2', 0): 'Gy',
				('Gzpi2',0):'Gz'
						}  # just for convenience, to make the labels prettier
for name,modelpack in zip(model_names,modelpacks):

	model = modelpack.target_model('GLND')
	gauge_basis = pygsti.baseobjs.CompleteElementaryErrorgenBasis(
					basis1q, model.state_space, elementary_errorgen_types='HSCA')
	model.setup_fogi(gauge_basis, None, op_abbrevs, include_spam=True, reparameterize=True)
	model.write(name)
```
